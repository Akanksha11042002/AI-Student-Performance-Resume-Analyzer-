```python
import re
import time
```

```python
skills_list = ["python", "sql", "excel", "machine learning", "communication"]
skills_set = set(skills_list)
skills_tuple = tuple(skills_list)
skills_dict = {"python": "intermediate","sql": "basic","excel": "basic","machine learning": "beginner","communication": "good"
}
print("Skills Loaded ")
```

```python
def resume_analyzer(resume):
    resume = resume.lower()
    found = list(filter(lambda skill: re.search(r'\b' + skill + r'\b', resume), skills_set))
    missing = list(skills_set - set(found))
    score = round((len(found) / len(skills_set)) * 100, 2)
    return found, missing, score
```

```python
def typing_test():
    sentence = "python is easy to learn"
    print("\nTyping Test")
    print("Type:", sentence)
    input("Press Enter to start...")
    start = time.time()
    typed = input("Start typing: ")
    end = time.time()
    time_taken = end - start
    words = len(typed.split())
    try:
        speed = round(words / (time_taken / 60), 2)
    except:
        speed = 0
    return speed
```

```python
def study_planner(missing):
    print("\nStudy Plan")
    if not missing:
        print("No missing skills ")
        return
    hours = int(input("Enter study hours per day: "))
    per_skill = hours // len(missing)
    for skill in missing:
        print(skill, "->", per_skill, "hrs/day")
```

```python
class Report:
    def __init__(self, score, speed):
        self.score = score
        self.speed = speed
    def generate(self):
        print("\nFINAL REPORT")
        if self.score >= 70 and self.speed >= 30:
            print("Excellent  Job Ready")
        elif self.score >= 40:
            print("Good  Improve More")
        else:
            print("Needs Improvement ")
        print("Score:", self.score, "%")
        print("Speed:", self.speed, "WPM")
```

```python
def main():
    print("AI Student Performance & Resume Analyzer \n")
    resume = input("Paste your resume:\n")
    found, missing, score = resume_analyzer(resume)
    print("\nFound Skills:", found)
    print("Missing Skills:", missing)
    print("Score:", score, "%")
    speed = typing_test()
    study_planner(missing)
    report = Report(score, speed)
    report.generate()
main()
```
